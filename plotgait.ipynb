```python
# Import library
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# define function to mapping plot filename
map_plot_filename = lambda filename: '_'.join(list(map(str.lower, filename.split(' '))))
```

```python
# Section 1 - Running accelerometer gait graph at x,y,z axis

# Import dataset
df = pd.read_excel('./acceleration.xls')
data = ['x', 'y', 'z']

# Plotting graph
for i in range(3):
    plot_name = f'Gait graph for linear acceleration data at {data[i]}-axis'
    plt.xlabel('time (s)')
    plt.ylabel('m/s^2')
    plt.title(plot_name)
    plt.plot(df['time'].to_numpy(), df[data[i]].to_numpy())
    plt.savefig(map_plot_filename(plot_name))
    plt.show()    
```

```python
# Section 2 - Running gyroscope gait graph at x,y,z axis

# Import dataset
df = pd.read_excel('./gyroscope.xls')
data = ['x', 'y', 'z']

# Plotting graph
for i in range(3):
    plot_name = f'Gait graph for gyroscope data at {data[i]}-axis'
    plt.xlabel('time (s)')
    plt.ylabel('rad/s')
    plt.title(plot_name)
    plt.plot(df['time'].to_numpy(), df[data[i]].to_numpy())
    plt.savefig(map_plot_filename(plot_name))
    plt.show()    
```
