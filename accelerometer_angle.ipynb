```python
# Import library
import pandas as pd
import math
import matplotlib.pyplot as plt
```

```python
# Section 1 - plotting and calculation accelerometer rho angle

# import dataset
df = pd.read_excel('./acceleration.xls')
print(df.head())
```

```python
# define accelerometer rho angle computation function
# with formula :
# arctan(Ax / sqrt(Ay^2 + Az^2))
rho_angle_func = lambda x,y,z: math.atan(x/ math.sqrt(math.pow(y, 2) + math.pow(z, 2)))
```

```python
# perform calculation with map and functional way
accelerometer_rho_angle = list(map(rho_angle_func, df['x'].to_numpy(), 
                                                   df['y'].to_numpy(), 
                                                   df['z'].to_numpy()))
```

```python
# define function to mapping plot filename
map_plot_filename = lambda filename: '_'.join(list(map(str.lower, filename.split(' '))))
```

```python
# plotting knee flexion based on time
plot_name = 'Accelerometer rho angle'
plt.xlabel('time (s)')
plt.ylabel('rad/s')
plt.title(plot_name)
plt.plot(df['time'].to_numpy(), accelerometer_rho_angle)
plt.savefig(map_plot_filename(plot_name))
plt.show()
```

```python
# ----------------------------------------------------------------------
# Section 2 - plotting gyroscope 
data = ['x', 'y', 'z']
for i in range(3):
    plot_name = f'Gait graph for gyroscope data at {data[i]}-axis'
    plt.xlabel('time (s)')
    plt.ylabel('rad/s')
    plt.title(plot_name)
    plt.plot(df['time'].to_numpy(), df[data[i]].to_numpy())    
    plt.savefig(map_plot_filename(plot_name))
    plt.show()    
```
